```python
import matplotlib.pyplot as plt
import numpy as np


# 1. Подготовка дискретных данных
x_range = np.arange(-5, 6)    # от -5 до 5 включительно
y_range = np.arange(-10, 11)  # от -10 до 10 включительно

# Создаем сетку координат
X, Y = np.meshgrid(x_range, y_range)

# 2. Ваша функция z = f(x, y)
Z = X**3 + 2 * X * Y + Y**3  # Пример: параболоид

# 3. Подготовка данных для bar3d (нужны плоские массивы)
x_flat = X.flatten()
y_flat = Y.flatten()
z_flat = Z.flatten()

# Основание каждого столбика (начинаем с нуля по оси Z)
bottom = np.zeros_like(z_flat)

# Размеры столбиков (ширина и глубина 0.8 для зазоров, либо 1.0 для плотного прилегания)
width = depth = 0.8 

# 4. Визуализация
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Отрисовка
bars = ax.bar3d(x_flat, y_flat, bottom, width, depth, z_flat, 
                shade=True, color='skyblue')

# Настройка осей
ax.set_xlabel('X (целые числа)')
ax.set_ylabel('Y (целые числа)')
ax.set_zlabel('Z = f(X, Y)')
ax.set_title('Дискретная поверхность (3D гистограмма)')

plt.show()
```
