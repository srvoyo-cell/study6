```python
import optimizers
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

Метод циклического покоординатного спуска \
Метод сопряженных направлений (метод Пауэлла) \
Метод градиентного спуска \
Метод наискорейшего спуска

```python
EPS_1 = 1e-5
EPS_2 = 1e-3
START_APPROX_1 = np.array([3, 10])
START_APPROX_2 = np.array([0, 0])

def func1(v):
    x, y = v
    return 24*x + 7*y - 4*x**2 - 6*y**2

def func2(v):
    x, y = v
    return 7*x - 0.8*y + np.exp(0.49*x**2 + 0.17*y**2)
```

```python
boundings_1 = [(-10, 10), (-10, 10)]
boundings_2 = [(-2, -1), (0, 1)]

x_1 = np.linspace(-10, 10, 1000)
y_1 = np.linspace(-10, 10, 1000)

x_2 = np.linspace(-2.1, 0.5, 1000)
y_2 = np.linspace(-2, 2, 1000)

X_1, Y_1 = np.meshgrid(x_1, y_1)
X_2, Y_2 = np.meshgrid(x_2, y_2)
Z_1 = func1([X_1, Y_1])
Z_2 = func2([X_2, Y_2])
```

```python
fig = make_subplots(rows=1, cols=2,
                    specs=[[{'is_3d': True}, {'is_3d': True}]],
                    )
fig.add_trace(go.Surface(x=X_1, y=Y_1, z=Z_1), 1, 1, )
fig.add_trace(go.Surface(x=X_2, y=Y_2, z=Z_2), 1, 2)
fig.show()
```

```python
extr_1 = optimizers.coordinate_descent(func1, START_APPROX_1, boundings_1,
                                        tol=EPS_1, mode='max')
extr_2 = optimizers.coordinate_descent(func2, START_APPROX_1, boundings_1, tol=EPS_1)
print(extr_1, extr_2)
```
